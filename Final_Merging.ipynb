```python
import pandas as pd
import random
```

```python
file_path = 'Airline Review/combined.csv'
review_factor_df = pd.read_csv(file_path, index_col=False)
```

```python
print(review_factor_df.shape)
review_factor_df.head()
```

```python
review_factor_df[['Month', 'Year']] = review_factor_df['date'].str.split(' ', expand=True)


review_factor_df['Month'] = pd.to_datetime(review_factor_df['Month'], format='%B').dt.month
review_factor_df['Year'] = review_factor_df['Year'].astype(int)
review_factor_df.insert(2, 'date', review_factor_df.pop('date'))
review_factor_df.insert(3, 'Year', review_factor_df.pop('Year'))
review_factor_df.insert(4, 'Month', review_factor_df.pop('Month'))
```

```python
review_factor_df['Quarter'] = pd.cut(review_factor_df['Month'], bins=[1, 4, 7, 10, 13], labels=[1, 2, 3, 4], right=False)
review_factor_df.insert(4, 'Quarter', review_factor_df.pop('Quarter'))
```

```python
review_factor_df = pd.read_csv('E:\zx/cleaned_review_factor.csv')
review_factor_df
```

```python
grouped = review_factor_df.groupby(['Airline', 'Year', 'Quarter']).size().reset_index(name='len')
result = grouped.to_dict(orient='records')
result
```

```python
zowie_df = pd.read_csv('E:\zx/zowie_combined.csv')
zowie_df.shape
```

```python
zowie_df = zowie_df.rename(columns={'TkCarrier': 'Airline'})
```

```python
zowie_df.head()
```

```python
# Iterate over each dictionary and remove rows randomly until the desired length is achieved
for attr in result:
    filtered_df = zowie_df[(zowie_df['Airline'] == attr['Airline']) & (zowie_df['Year'] == attr['Year']) & (zowie_df['Quarter'] == attr['Quarter'])]
    if len(filtered_df) > attr['len']:
        remove_indices = random.sample(list(filtered_df.index), len(filtered_df) - attr['len'])
        zowie_df.drop(remove_indices, inplace=True)
```
